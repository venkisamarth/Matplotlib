# Cutomizing the Line Plot

```python
#importing the libraries
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#defining the dataset
x = np.arange(0,10,1)
y = 2*x + 5
#changing the figure
fig=plt.figure(figsize=(10,5))
#plotting the datapoints
plt.plot(x,y,linewidth = 2.0,linestyle = ":", color = 'r',alpha = 0.5,marker ='o')
plt.title("Line Plot Demo")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.legend(['line1'],loc='best')
plt.grid(True)
plt.show()
```


Line plots, also known as line charts or line graphs, are used to display the relationship between two continuous variables. They are particularly useful for visualizing trends over time or continuous data points. Here are some common conditions in which line plots are used:


1 Time-series data: Line plots are often employed to showcase changes in a variable over time. This could include tracking stock prices over weeks or months, monitoring temperature variations throughout the day, or observing population growth over years.

2 Continuous data: Line plots are suitable for representing continuous data points, such as measurements taken at regular intervals. For example, you can use a line plot to illustrate the height of plants at different time points in an experiment or to display the progression of a person's weight loss journey over several months.

3 Comparisons between groups: Line plots can be used to compare trends or patterns between different groups. By plotting multiple lines on the same graph, you can easily compare the changes over time for different categories or subsets of data.

4 Correlations and relationships: Line plots can help visualize the relationship between two continuous variables. By plotting one variable on the x-axis and another on the y-axis, you can assess if there is a positive or negative correlation between them and observe any patterns or trends.

5 Forecasting and predictions: Line plots can be used to project future trends based on historical data. By extending the line beyond the available data points, you can make predictions or estimates for future values.


 subplot():

 In Matplotlib, a subplot refers to a grid of individual plots within a single figure. It allows you to create multiple plots or axes within the same window, facilitating the visualization of multiple graphs or data comparisons in a structured manner.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x = np.arange(0,10,1)
y1 = 2*x+5
y2 = 3*x+10

#Create two or more plots in a fig
plt.subplot(1,2,1)
#PLotting the points
plt.plot(x,y1)
plt.title('Graph1')

plt.subplot(1,2,2)
plt.plot(x,y2)
plt.title('Graph2')
plt.show()
```

# How to create a Bar Plot

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#creating the dataset
data = {'apples':20,'Mangoes':15,'Lemon':30,'Oranges':10}
names = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10,5))
#Creating the bar plot
plt.bar(names,values)
plt.show()
```

Bar plots:


  A bar plot, also known as a bar chart or bar graph, is a type of data visualization that uses rectangular bars to represent and compare categorical data. It displays the values of different categories as bars with lengths proportional to the values they represent.

Bar plots are commonly used in the following conditions:

1  Comparing categorical data: Bar plots are ideal for comparing the values of different categories or groups. They are suitable for visualizing data that can be divided into distinct categories, such as comparing sales figures for different products, comparing student performance across different subjects, or comparing the population of different countries.

2   Frequency or count data: Bar plots are often used to display the frequency or count of different categories. For example, representing the number of customers in different age groups, the frequency of different diseases in a population, or the number of occurrences of specific events.

3  Ranking or ordering data: Bar plots can be used to represent data in a ranked or ordered manner. They allow for clear visualization of the hierarchy or order of different categories. For instance, displaying the top 10 countries with the highest GDP, ranking the popularity of different programming languages, or showcasing the performance of different athletes in a competition.

4  Comparing data over time: Bar plots can be used to visualize changes in data over time for different categories. By using stacked or grouped bars, you can compare the values of different categories at different time points. For example, tracking monthly sales figures for multiple products, comparing the market share of different companies over several years, or examining the distribution of different car models sold each month.

5  Visualizing proportions or percentages: Bar plots can represent the proportion or percentage distribution of different categories. By normalizing the data, the bar lengths can indicate the relative proportions rather than the actual values. This is useful for visualizing market share, demographic distributions, or survey responses.

6  Bar plots are versatile and effective for comparing and visualizing categorical data. They provide a clear representation of the relationships, differences, and distributions between categories, making them useful for various analytical and exploratory purposes.







# Customizing the Bar Chart

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#creating the dataset
data = {'apples':20,'Mangoes':15,'Lemon':30,'Oranges':10}
names = list(data.keys())
values = list(data.values())

fig = plt.figure(figsize = (10,5))
#Creating the bar plot
plt.barh(names,values,color="orange")
plt.title("Bar Graph Demo")
plt.xlabel("Fruits")
plt.ylabel("Quantity")
plt.show()
```

# scatter plot


  A scatter plot is a type of data visualization that displays the relationship between two continuous variables. It uses dots or markers to represent individual data points and shows how the values of the two variables are related to each other.

  Scatter plots are commonly used in the following conditions:

1  Correlation analysis: Scatter plots are useful for assessing the relationship or correlation between two variables. They can help determine if there is a positive, negative, or no correlation between the variables. For example, you can use scatter plots to analyze the relationship between a person's age and their cholesterol level, or the relationship between study time and exam scores.

2  Clustering and grouping: Scatter plots can aid in identifying clusters or groups within the data. By plotting the data points, you can visually assess if there are natural groupings or patterns. This is particularly helpful in exploratory data analysis or data mining tasks.

3  Outlier detection: Scatter plots can reveal outliers, which are data points that significantly deviate from the overall pattern. Outliers can be identified as data points that fall far away from the majority of the points in the scatter plot. Detecting and analyzing outliers can provide insights into unusual or exceptional observations within the data.

4  Trend visualization: Scatter plots can show trends or patterns in the data. By observing the general direction or shape of the scatter plot, you can identify trends such as linear relationships, quadratic relationships, or exponential relationships between the variables.

5  Comparison of multiple groups: Scatter plots can be used to compare the relationships between two variables across different groups or categories. By using different colors or markers for each group, you can visually compare the patterns or correlations within each group.

6  Regression analysis: Scatter plots are often employed as a preliminary step in regression analysis. They can help visualize the relationship between the independent and dependent variables and assist in determining if a linear regression model is appropriate for the data.

7  Scatter plots provide a visual representation of the relationship between two continuous variables and are useful for identifying patterns, correlations, outliers, and trends within the data. They are versatile tools in data analysis and visualization and are especially valuable in exploratory data analysis.

# Customizing our Scatter Plot

```python
import matplotlib.pyplot as plt
%matplotlib inline

#Creating the dataset
a =[10,20,30,40,50,60,70,80]
b =[5,3,2,5,6,1,4,2]
x =[1,2,3,4,3,2,1,2]

#Creating the scatter Plot
plt.scatter(a,b, c='g', s=200,edgecolors='y',marker='o',alpha =0.5)
plt.scatter(a,x,c='r', s=400,edgecolors='b',marker='4',alpha =1)
plt.legend(['b','x'],loc='best')
plt.title("Scatter Plot Demo")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
plt.savefig('Scatter Plot.png')
plt.show()
```


A histogram is a graphical representation of the distribution of numerical data. It displays the frequencies or counts of data points falling within different intervals, or "bins," along the range of the data. Histograms are particularly useful in the following conditions:

1 Data distribution analysis: Histograms are commonly used to examine the distribution of data. They provide a visual representation of how the data is spread out and can reveal patterns such as symmetry, skewness, or multimodality. Histograms allow you to see if the data is normally distributed, skewed to the left or right, or has other distinct characteristics.

2 Frequency or count analysis: Histograms show the frequency or count of data points within each bin. This is useful for understanding the frequency distribution of data and identifying which intervals contain a larger or smaller number of data points. Histograms can be used to analyze the occurrence of events, the distribution of exam scores, or the frequency of different outcomes.

3 Range and outliers detection: Histograms help identify the range and spread of data values. By observing the width and shape of the histogram, you can estimate the minimum and maximum values and detect potential outliers. Outliers are data points that fall outside the expected range or distribution and are often represented as bars that stand out from the majority of the histogram.

4 Data comparison: Histograms allow for the visual comparison of different datasets or groups. By overlaying multiple histograms or using different colors, you can compare the distributions of variables across different categories or subsets of data. This is useful for analyzing the differences in income distributions between different countries, comparing the heights of individuals from different regions, or assessing the performance of different products in terms of sales.

5 Probability density estimation: Histograms can be used to estimate the probability density function of a dataset. By normalizing the histogram bars to represent the relative frequencies or probabilities, you can obtain an approximation of the underlying probability distribution. This can be useful for modeling or understanding the data distribution and for making probabilistic inferences.

6 Histograms provide a visual representation of the distribution and frequency of data points, making them valuable for analyzing data distributions, detecting outliers, comparing datasets, and estimating probability densities. They are widely used in data analysis, exploratory data analysis, and statistical inference.

A box plot, also known as a box-and-whisker plot, is a data visualization technique that displays the summary statistics of a dataset, such as the minimum, maximum, median, quartiles, and outliers. It provides a concise summary of the distribution and variability of the data.


Box plots are commonly used in the following conditions:

1 Distribution comparison: Box plots are effective for comparing the distributions of multiple datasets or groups. By placing multiple box plots side by side, you can visually compare the central tendency, spread, and skewness of different groups or categories. This is useful for analyzing the distribution of a variable across different populations, treatments, or experimental conditions.

2 Outlier detection: Box plots can help identify outliers, which are data points that significantly deviate from the overall distribution. The whiskers in a box plot extend to the minimum and maximum values within a certain range (often determined by 1.5 times the interquartile range). Individual data points lying outside this range are considered outliers and can be visually represented in the plot. Detecting outliers is valuable for identifying unusual or extreme observations within the dataset.

3 Skewness and symmetry assessment: Box plots provide insights into the skewness and symmetry of the data. The position of the median line within the box indicates the central tendency, while the length and symmetry of the box reveal the spread and skewness. Skewness refers to the asymmetry of the data distribution, and box plots can visually indicate if the data is skewed to the left or right.

4 Variability analysis: Box plots display the interquartile range (IQR), which represents the middle 50% of the data. The length of the box indicates the spread of the data within this range, allowing you to assess the variability or dispersion of the dataset. Comparing the lengths of the boxes between different groups can provide insights into differences in variability.

5 Data summary visualization: Box plots present a compact visual summary of the dataset's key statistics, including the minimum, maximum, median, and quartiles. This allows for a quick understanding of the range, central tendency, and spread of the data without examining individual data points.

6 Box plots are valuable tools for summarizing and comparing the distributions of datasets, identifying outliers, assessing skewness and variability, and presenting key statistics in a concise manner. They are widely used in exploratory data analysis, statistical analysis, and data presentation.






# How to create a Box Plot?

```python
import matplotlib.pyplot as plt
%matplotlib inline

#Data Preparation
total = [20,4,1,30,20,10,20,70,30,10]
order = [10,3,1,15,17,2,30,44,2,1]
discount = [30,10,20,5,10,20,50,60,20,45]
data = list([total, order,discount])
#Plotting the data
plt.boxplot(data, showmeans = True)
plt.title("Box Plot Demo")
plt.grid(True)
plt.show()
```


A violin plot is a data visualization technique that combines aspects of a box plot and a kernel density plot. It provides a more detailed representation of the distribution of data compared to a traditional box plot.

A violin plot is commonly used in the following conditions:

1  Distribution comparison: Violin plots are effective for comparing the distributions of multiple datasets or groups. Like a box plot, a violin plot displays the summary statistics such as the median, quartiles, and outliers. However, it also includes a kernel density estimation on each side of the box, which provides a smooth representation of the data distribution. By comparing the widths and shapes of the violin plots, you can visually assess differences in the distribution of variables across different categories or groups.

Skewness and symmetry assessment: Violin plots provide insights into the skewness and symmetry of the data distribution. The width of the violin at different points indicates the density or frequency of the data values. If the distribution is symmetric, the violin plot will have a similar width on both sides of the median line. If the distribution is skewed, the width will vary, indicating the concentration of data points at different values.

2  Kernel density estimation: Violin plots include kernel density estimates, which provide a smoothed representation of the data distribution. This can be helpful for understanding the shape and smoothness of the distribution, identifying multiple peaks or modes, and detecting potential outliers. The density plot inside the violin allows for a more detailed visualization of the distribution compared to a traditional box plot.

3  Outlier detection: Similar to a box plot, violin plots can help identify outliers. Outliers are represented as individual points outside the range of the whiskers or violin plot. By observing the extent of the whiskers or the presence of outliers, you can detect extreme or unusual values within the dataset.

4  Violin plots provide a comprehensive visualization of the distribution, central tendency, variability, skewness, and density estimation of data. They are particularly useful for comparing distributions, assessing skewness, and providing a detailed representation of the data distribution beyond the summary statistics of a box plot.






# How to create a violin plot?

```python
import matplotlib.pyplot as plt
%matplotlib inline

#Data Preparation
total = [20,4,1,30,20,10,20,70,30,10]
order = [10,3,1,15,17,2,30,44,2,1]
discount = [30,10,20,5,10,20,50,60,20,45]
data = list([total, order,discount])
#Plotting the data
plt.violinplot(data, showmeans = True, showmedians = True)
plt.title("Violin Plot Demo")
plt.grid(True)
plt.show()
```

A pie chart is a circular data visualization that displays the proportions or percentages of different categories within a whole. The chart is divided into sectors, where each sector represents a specific category, and the size of the sector corresponds to the proportion or percentage it represents.

Pie charts are commonly used in the following conditions:

 1  Proportional comparison: Pie charts are useful for visualizing the relative proportions of different categories within a whole. Each category's share is represented by the angle or area of its corresponding sector. Pie charts make it easy to compare the magnitudes of different categories and identify the most dominant or significant categories.

2 Composition analysis: Pie charts are effective for showing the composition or breakdown of a whole into its constituent parts. This is particularly useful when you want to understand the contribution of different categories to the total or analyze the distribution of a variable among its components. For example, representing the expenditure breakdown of a budget, showing the market share of different companies, or illustrating the distribution of population by age groups.

3  Percentage visualization: Pie charts visually represent percentages, allowing for a quick understanding of the distribution or composition of a dataset. Each sector's size represents the proportion it contributes to the whole, making it easy to compare the relative sizes of different categories.

4  Limited number of categories: Pie charts work best when representing a small number of categories. As the number of categories increases, the individual sectors become smaller and harder to distinguish, making it challenging to accurately interpret the chart.

5  Simplified data visualization: Pie charts provide a simplified and intuitive representation of data. They are easy to interpret, even for individuals who are not familiar with statistical or numerical analysis.

6  It is important to note that while pie charts have their uses, they are also subject to certain limitations. For example, they may not be suitable for displaying precise numerical values or comparing small differences between categories. In such cases, alternative visualizations like bar charts or stacked column charts may be more appropriate.

7  Pie charts are effective for visually representing the proportions, composition, and percentages of different categories within a whole. They are widely used in presentations, reports, and dashboards to communicate information in a visually appealing and easily understandable manner.

# How to create a Pie Chart



```python
#importing the required libraries
import matplotlib.pyplot as plt
%matplotlib inline

#preparing the dataset
label = ['Dog','Cat','Wolf','Lion']
sizes = [50,45,60,80]

#creating the pie chart
plt.pie(sizes, labels = label,autopct='%1.1f%%',shadow=True, startangle = 90,explode = (0,0.1,0,0))
plt.title("Pie Chart Demo")
plt.show()
```

A doughnut chart is a variant of the traditional pie chart that has a hole in the center, giving it a "doughnut" shape. It shares similarities with the pie chart in terms of visualizing proportions or percentages, but it also offers some distinct features.

Doughnut charts are commonly used in the following conditions:

1 Proportional comparison: Like pie charts, doughnut charts are useful for comparing the proportions or percentages of different categories within a whole. Each category is represented by a sector, and the size of the sector indicates its relative proportion. Doughnut charts allow for a quick visual comparison of the magnitudes of different categories.

2 Composition analysis: Doughnut charts effectively show the composition or breakdown of a whole into its constituent parts. They provide a clear visualization of the distribution of a variable among its components, similar to pie charts. The inner and outer rings of the doughnut chart can represent different levels of categorization or hierarchical structures, allowing for a more detailed analysis of the data.

3 Multiple datasets or comparisons: Doughnut charts can be used to compare multiple datasets or make side-by-side comparisons. By placing multiple rings or sets of rings within the chart, you can visualize the proportions of different categories across different datasets or groups. This is particularly useful for comparing data from different time periods, different regions, or different subgroups.

4 Emphasizing the whole: The central hole in the doughnut chart can be used to draw attention to the whole or total. By presenting the total value or the overall proportion in the center, it becomes more prominent and easier to perceive.

5Limited number of categories: As with pie charts, doughnut charts work best when representing a small number of categories. Having too many categories can lead to cluttered and confusing visuals, as the individual sectors become smaller and harder to distinguish.

6  Doughnut charts offer a visually appealing alternative to pie charts, providing a way to compare proportions, analyze composition, and make multiple comparisons. However, it's important to consider the limitations of doughnut charts, such as the potential for imprecise comparisons and difficulties in accurately perceiving the sizes of individual sectors. Depending on the specific data and goals, other types of charts like stacked bar charts or treemaps may be more suitable alternatives.






# How to create a Doughnut Chart

```python
import matplotlib.pyplot as plt
%matplotlib inline

#define the dataset
group_names = ["GroupA","GroupB","GroupC"]
group_size = [20,30,50]
size_centre = [5]
a,b,c = [plt.cm.Blues,plt.cm.Reds,plt.cm.Greens]
#Creating the doughnut chart
pie1 = plt.pie(group_size,labels = group_names, radius = 1.5,colors = [a(0.5),b(1.5),c(1.0)])
pie2 = plt.pie(size_centre,radius = 1.0,colors = 'w')
plt.show()
```

An area chart is a data visualization technique that represents the quantitative values of different categories or variables using filled areas. It is similar to a line chart, but the area between the line and the horizontal axis is filled with color or pattern, creating a visual representation of the cumulative values.

Area charts are commonly used in the following conditions:

1  Tracking trends over time: Area charts are particularly effective for visualizing the changes in quantitative values over time. By plotting the values on a continuous time axis and connecting them with lines, the filled areas beneath the lines provide a visual representation of the accumulated values. This allows for easy tracking of trends, fluctuations, and patterns in the data.

2  Comparative analysis: Area charts can be used to compare multiple categories or variables. By plotting multiple filled areas on the same chart, it becomes easy to observe the relative changes and comparisons between the categories. This is useful for analyzing the performance of different products, the distribution of revenue among different business segments, or the comparison of population trends across different regions.

3  Part-to-whole relationships: Area charts are suitable for illustrating part-to-whole relationships within a dataset. By stacking the filled areas on top of each other, the overall area represents the total value or 100% of the dataset, while the individual areas represent the proportions or contributions of different categories. This is useful for visualizing the market share of different competitors, the composition of a budget, or the distribution of sales across different product categories.

4  Highlighting cumulative data: Area charts are effective in emphasizing the cumulative values over time or across different categories. The filled areas beneath the lines create a visual emphasis on the accumulated values, making it easier to interpret and compare the total values between different points or categories.

5  Visualizing uncertainty or range: Area charts can also be used to represent the range or uncertainty of data. By using transparent or shaded areas, you can visually indicate the lower and upper bounds of a range. This is useful for representing statistical measures such as confidence intervals or uncertainty in forecasting models.

6  Area charts provide a visually appealing and intuitive way to represent trends, compare categories, analyze part-to-whole relationships, and highlight cumulative data. They are commonly used in time series analysis, financial data analysis, and market analysis, among other fields.

# How to create a Area Chart?

```python
import matplotlib.pyplot as plt
%matplotlib inline

#define the dataset
x=range(1,15)
y=[1,4,6,8,4,5,3,2,4,1,5,6,8,7]

#creating the area plot
plt.stackplot(x,y,colors = 'g', alpha = 0.5)
#Merge area chart with the line chart
plt.plot(x,y,color = 'g')
plt.grid(True)
plt.show()
```
