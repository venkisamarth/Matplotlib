```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.image as mpimg
#check version
print(pd.__version__)
print(np.__version__)
```

```python
fig,ax=plt.subplots()
x=np.linspace(0,10,300)
y=np.sin(x)+x
ax.plot(x,y)
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(10,4))#fig,plt.figure(figsize=(10,4))
fig.suptitle('our first plot',fontsize=20)
ax.plot(x,y,color='r',ls='--',linewidth=4)
```

```python
fig,ax=plt.subplots(figsize=(10,5))
fig.suptitle('our first plot',fontsize=40)
ax.plot(x,y,color='b',ls='dashed',linewidth=5)
ax.set_ylabel('function',fontsize=20)#for change the label
ax.set_xlabel('argument',fontsize=20)
#ax.set_xlim(1,20)
#ax.set_ylim(1,20)  for change the scale of the figure
```

```python
#subplot

fig,axes=plt.subplots(1,3,figsize=(16,8))
ax1,ax2,ax3=axes
x=np.linspace(0,10,500)
y1=np.sin(x)
y2=np.cos(x)
y3=np.tan(x)

ax1.set_ylabel('sin(x)',fontsize=30)
ax1.set_ylabel('x',fontsize=30)

ax2.set_ylabel('cos(x)',fontsize=30)
ax2.set_xlabel('x',fontsize=30)

ax3.set_ylabel('tan(x),',fontsize=30)
ax3.set_xlabel('x',fontsize=30)
fig.tight_layout(pad=2)


ax1.plot(x,y1,c='r',linewidth=4)
ax2.plot(x,y2,c='b',linewidth=4)
ax3.plot(x,y3,c='g',linewidth=4)
```

```python
years=[1950,1955,1960,1965,1970,1980,1985,1990,1995,2000,2005,2010,2015]
population=[2.5,2.7,3.0,3.3,3.6,4.0,4.4,4.8,5.3,5.7,6.1,6.5,7.3]
plt.plot(years,population,color='b')
plt.xlabel('population in Billion')
plt.ylabel('population growth by year')
plt.title('population growth')
```
